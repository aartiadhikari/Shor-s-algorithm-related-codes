```python
import numpy as np
import matplotlib.pyplot as plt

# Range of input sizes (number of bits)
n_bits = np.linspace(100, 2000, 200)

# Number Field Sieve runtime (sub-exponential)
# Formula: exp(1.9 * (log n)^(1/3) * (log log n)^(2/3))
nfs_runtime = np.exp(1.9 * (np.log(n_bits) ** (1/3)) * (np.log(np.log(n_bits)) ** (2/3)))

# Shor's algorithm runtime (polynomial: (log n)^3)
shor_runtime = (np.log(n_bits) ** 3)

# Normalize for visualization (to keep scale readable)
nfs_runtime_scaled = nfs_runtime / max(nfs_runtime) * 400
shor_runtime_scaled = shor_runtime / max(shor_runtime) * 50

# Plotting
plt.figure(figsize=(9,5))
plt.plot(n_bits, nfs_runtime_scaled, label="Number Field Sieve\nExp(1.9(log n)^(1/3)(log log n)^(2/3))", color="blue")
plt.plot(n_bits, shor_runtime_scaled, label="Shor's runtime (log n)^3", color="green")

plt.xlabel("N-BITS", fontsize=12)
plt.ylabel("RUN TIME", fontsize=12)
plt.title("Runtime Comparison: Number Field Sieve vs Shor's Algorithm")
plt.legend()
plt.grid(True)
plt.show()
```
