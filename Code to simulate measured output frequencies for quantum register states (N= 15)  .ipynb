```python
%matplotlib inline
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

# Define registers
n = 4  # input register (x)
m = 4  # output register (y)

qreg_x = QuantumRegister(n, 'x')
qreg_y = QuantumRegister(m, 'y')
creg = ClassicalRegister(n + m, 'c')  # classical register for all qubits

# Create the quantum circuit
qc = QuantumCircuit(qreg_x, qreg_y, creg)

# Apply Hadamard gates to input register
qc.h(qreg_x)

# Measure all qubits
qc.measure(qreg_x[:] + qreg_y[:], creg[:])

# Simulate using AerSimulator
sim = AerSimulator()

# âœ… Transpile the circuit for the simulator
compiled_circuit = transpile(qc, backend=sim)

# Run simulation
result = sim.run(compiled_circuit, shots=1024).result()

# Plot the result
counts = result.get_counts()
plot_histogram(counts, title="Figure 1: Quantum Register Initialization for N=15")
plt.show()
```
